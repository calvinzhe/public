### Setting up Jupyter Notebook
This setup uses the IPython kernel, but Jupyter Notebook can be used to run a number of different kernels like R.  The easiest way to set up your computer to run Jupyter for this purpose is to download and install Anaconda.  However, if you need to set up Jupyter from the base Python installation, you will have to install the following modules:

- Jupyter
- scipy modules:
     - ipython
     - sympy
     - numpy
     - matplotlib
     - pandas

You can do so using the command:

    pip install [module]

on the command prompt (Windows) or terminal (Linux) once you have Python installed, replacing [module] with the name of the module you are trying to install.  After installing the necessary packages, the Jupyter Notebook can be started by entering the command:

    jupyter notebook
    
on the command prompt or terminal.  The web interface will load and from there, you can create, delete, and open notebooks in your computer's directory.

Jupyter Notebooks (.ipynb) can be saved to a github repository and viewed online using nbviewer: https://nbviewer.jupyter.org/. Jupyter Notebooks can also be exported to formats such as HTML and posted as a webpage, however, I've yet to get .pdf exporting to work.

## Math

Some Symbolic math examples using SymPy.  Unfortuantely, it looks like you are only able to print one symbolic math output per cell.

```python
x_1 = symbols('x_1')
f = x**2 + 5*x + cos(x) + atan(x) + tanh(x) + x_1
f
```

```python
g = diff(f,x)
g
```

```python
h = Function('h')
F = Eq(h(x).diff(x)+h(x), sin(x))
F
```

## Data

Generating a mock data set.  You can use this method (with high enough resolution) to generate theoretical curves.  With real data, you would use Pandas or NumPy to load float arrays into the kernel.

```python
xmin = 0
xmax = 10
xstep = 0.25 #Determines distance between adjacent points and hence the resolution of the graph array
dat1 = [i*xstep+xmin for i in range(int(xmax/xstep))]
dat2 = [40*sin(i) for i in dat1]
dat3 = [i**2 for i in dat1]
```

## Plotting

The first plot is of the mock data plotted using matplotlib.  The latter plot are of SymPy functions using the SymPy plotting tools.

```python
plt.title("Title")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(0,10)
plt.ylim(-40,40)
plt.plot(dat1, dat2, 'g.')
plt.plot(dat1, dat3, 'r.')
```
